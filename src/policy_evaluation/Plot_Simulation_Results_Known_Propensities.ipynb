```python
import sys
# sys.path.insert(0, '../ATE_Simulations/Results')
sys.path.insert(0, '../../src')
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
import os
from visualization_utils import perc, SetPlotRC, ApplyFont

import warnings
warnings.filterwarnings("ignore")

# plt.rcParams['text.usetex'] = True
```

# Number of Observations Experiments

```python
full_results1 = pd.read_csv(r"Results/OPE_n_observations_result_100_observations.csv")
full_results2 = pd.read_csv(r"Results/OPE_n_observations_result_1000_observations.csv")
full_results3 = pd.read_csv(r"Results/OPE_n_observations_result_2000_observations.csv")
full_results4 = pd.read_csv(r"Results/OPE_n_observations_result_5000_observations.csv")
# full_results5 = pd.read_csv(r"Results/OPE_n_observations_result_10000_observations.csv")
full_results = pd.concat([full_results1, full_results2, full_results3, full_results4])

full_results.loc[full_results['Estimator'] =='ips_estimator', 'Estimator'] = 'IPS'
full_results.loc[full_results['Estimator'] =='Direct', 'Estimator'] = 'DM - NN'
full_results.loc[full_results['Estimator'] =='doubly robust estimator', 'Estimator'] = 'DR - NN'
full_results.loc[full_results['Estimator'] =='cmebis', 'Estimator'] = 'DM - CMPE'
full_results.loc[full_results['Estimator'] =='dr_bis', 'Estimator'] = 'DR-CMPE'
SetPlotRC()
# # # Plotting results
# sns.set(style="whitegrid")
# plt.figure(figsize=(8, 5))
# sns.boxplot(
#     data=full_results, x="Observation Size", y="MSE", hue="Estimator"
#     # , marker="o"
# )
# plt.yscale("log")
# plt.title("Number of Observations vs MSE")
# plt.tight_layout()
# ApplyFont(plt.gca(), xlabel_text_size = 15.0, ylabel_text_size = 15.0, title_text_size = 15.0, ticks_text_size = 15,)
# plt.show()

# Plotting results
sns.set(style="whitegrid")
plt.figure(figsize=(8, 5))
sns.pointplot(
    data=full_results, x="Observation Size", y="MSE", hue="Estimator",
    linestyles=["--", ":",":", "-.", "-"], markers=['o','v','x','^','+']
)
plt.yscale("log")
plt.title(r"Number of Observations vs MSE, $N = 50$, $K = 4$, $M = 20$, $d = 10$, $\alpha = -0.3$")
plt.xlabel("Number of Observations (n)")
plt.tight_layout()

ApplyFont(plt.gca(), xlabel_text_size = 15.0, ylabel_text_size = 15.0, title_text_size = 15.0, ticks_text_size = 15,)
plt.savefig('Figures/OPE_n_observations_experiment_known_props.png', format='png', bbox_inches='tight')
plt.show()
```
